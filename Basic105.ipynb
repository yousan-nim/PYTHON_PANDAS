```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_used_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_used_python
```

```python
python_df = pd.concat([country_respondents, country_used_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondent', 
                          'LanguageWorkedWith' : 'NumKnowsPython'})
python_df
```

```python
python_df['PctKnowsPython'] = ( python_df['Country'] / python_df['LanguageWorkedWith']) * 100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
